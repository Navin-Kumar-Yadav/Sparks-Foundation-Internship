```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SampleSuperstore.csv",encoding="unicode_escape")
```

## Different Ship Modes Present

```python
ship_modes = df['Ship Mode']
graph = sns.countplot(data=df, x=ship_modes, hue="Ship Mode",palette='flare')
```

## Unique Segments 

```python
unique_segments = df['Segment']
graph = sns.countplot(data=df,x=unique_segments,hue='Segment',palette='rocket_r')
```

## Countries Present in the dataset

```python
countries = df['Country']
graph = sns.countplot(data=df,x='Country',color='lightblue')
```

## Range of Postal Codes

```python
postal_code_min,postal_code_max = df['Postal Code'].min(),df['Postal Code'].max()
postal_code_min,postal_code_max
```

## Top 10 Highest Sales Cities

```python
cities_sales_highest = df.groupby('City',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False).head(10)
plt.xticks(rotation='vertical')
graph = sns.barplot(data=cities_sales_highest,x='City',y='Sales',hue='City',palette='viridis')
plt.tight_layout()
```

## Top 10 Lowest Sales Cities

```python
cities_sales_lowest = df.groupby('City',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=True).head(10)
plt.xticks(rotation='vertical')
graph = sns.barplot(data=cities_sales_lowest,x='City',y='Sales',hue='City',palette='crest')
plt.tight_layout()
```

## States covered in the dataset

```python
states = df["State"].unique()
plt.figure(figsize=(15,5))
plt.bar(x=states,height=.5)
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```

## Regions represented in the data

```python
graph = sns.countplot(data=df,x='Region',hue='Region',palette='viridis')
```

## Categories of products available in the dataset

```python
graph = sns.countplot(data=df,x='Category',hue='Category',palette='viridis')
for i in graph.containers:
    graph.bar_label(i)
```

## Sub-Categories present in the dataset

```python
sub_categories = df.groupby(["Sub-Category"], as_index = False)["Quantity"].sum().sort_values(by = "Quantity", ascending = False)
graph = sns.barplot(x = "Sub-Category", y = "Quantity", data= sub_categories)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.tight_layout()
for i in graph.containers:
    plt.bar_label(i)
```

## Total Sales recorded in the dataset

```python
total_sales = df['Sales'].sum()
total_sales
```

## Units of products sold in total

```python
products_sold = df.Quantity.sum()
products_sold
```

## Average Discount offered on products

```python
average_discount = df.Discount
average_discount.mean()*100
```

## Most and Least Profitable entries in the dataset

```python
min_profit = df.Profit.min()
max_profit = df.Profit.max()
min_profit, max_profit
```

## Correlation between quantity sold and profit

```python
quantity_sales = df.groupby(["Sub-Category"], as_index = False)[["Quantity", "Profit"]].sum().sort_values(by = "Profit", ascending = False)
graph = sns.barplot(x = "Quantity", y = "Profit", data=quantity_sales,hue='Sub-Category')
for bars in graph.containers:
    plt.bar_label(bars)
```

## Distribution of Sales across differentregions

```python
region_sales = df.groupby('Region',as_index=False)['Sales'].sum().sort_values(by='Sales',ascending=False)
graph = sns.barplot(x = "Region", y = "Sales", data=region_sales,hue="Region")
for bars in graph.containers:
    plt.bar_label(bars)
```
